```python
with open('morningstar_fundamentals_text.txt','r') as f: 
    morningstar = f.read()
```

```python
morningstar_factors = morningstar.split('\n')
morningstar_factors = [factor for factor in morningstar_factors if len(factor) > 0]
morningstar_factors = [factor for factor in morningstar_factors if len(factor.split(' ')) == 1]
print(morningstar_factors)
```
